# Create random 100 tosses and save it into a csv file!!!

```python
# Generate 100 random coin flips
# Track total heads and total tails each flip
# Write a CSV file with a row for each flip
import csv
import random 

#open a csv
with open("flips.csv","w") as file:
    writer = csv.writer(file, lineterminator = "\n")
    
    total_heads =0
    total_tails =0
    
    header = ['Number_of_flips','Outcome','Total_Header','Total_Tails']
    writer.writerow(header)
    
    for i in range(1,101,1):
        result= random.randrange(0,2)
        
        if result == 1:
            total_heads+=1
            output=[i,"Heads", total_heads,total_tails]
            writer.writerow(output)
            
        else:
            total_tails+=1
            output=[i,"Tails", total_heads,total_tails]
            writer.writerow(output)
#Add a completion message to let the user know that the operation has been completed
print("Complete!")
```
