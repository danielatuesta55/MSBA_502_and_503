```python
import tkinter 
import random 
import tkinter.messagebox
import tkinter.ttk

def label_hover(event):
    """Show an information label when hovering """
    global root
    new_label = tkinter.Label(root,texr='You have an 80% chance to win')
    new_label.configure(bg='yellow')
    new_label.configure(font="Arial 10 italic bold")
    new_label.grid(row=1, column=0)

def label_hover_end():
    global new_label
    new_label.grid_remove()
    
def get_fortune():
    """Gives a positive number 80% of the time nad negative 20% """
    fortune = random.randrange(1,6)
    if fortune <= 4:
        tkinter.messagebox.showinfo('Yay, you will be fortunate today')
    else:
        tkinter.messagebox.showwarning('Big yikes!, Your luck is crap')

root = tkinter.Tk()
root.title("Fortune teller")

info_label = tkinter.Label(root, text = "Click on the button to get fortune!!!")
info_label.grid(row=0,column=0)
info_label.bind("<Enter>",label_hover)
info_label.bind("<Leave>",label_hover_end)

fortune_button = tkinter.Button(root,text="Show furtune", command=get_fortune)
fortune_button.configure(fg='green')
fortune_button.grid(row=0,column=1)

root.mainloop()
#Create a "Furtone Teller"

```
