```python
#Import dependencies 
import tkinter
import tkinter.messagebox

def show_result():
    '''Show message box with the soccer results'''
    tkinter.messagebox.showinfo("Soccer results","Real Madrid wins!")
    
#Create a window
root = tkinter.Tk() 
root.title("Soccer")
root.configure(bg='light green')

madrid_lable= tkinter.Label(root,text="Real Madrid: 3")
madrid_lable.configure(bg='light green')
madrid_lable.pack()

pool_lable= tkinter.Label(root,text="Liverpool: 1")
pool_lable.configure(bg='light green')
pool_lable.pack()

result_button = tkinter.Button(root,text='Click me!',command = show_result)

result_button.pack()

root.mainloop()
```

# Placing widgets in tkinter

- Last class, we used .pack() to place wach widget in with tkinter
- The .pack() method stacks each widget vertically by default
- The most recent widget 
- An alternative methof to .pack() is to use .grid() this method allows you to specify which row and coloum to place each widget 
- For example new_label.grid(row=0,column=1) would plave an element in the bkue square below

```python
#import dependencies
#Create two buttons using grid() one that says hello and the other one that says goodbye

import tkinter
import tkinter.messagebox

def hello_fun():
    """Show message box"""
    tkinter.messagebox.showinfo('Hello',"Hello!!!!!!! \n Hello!!!!!!!")
def goodbye():
    '''Show goodbye function'''
    tkinter.messagebox.showinfo('Goodbye','Goodbye!!!!!')

#Create a window
root = tkinter.Tk() 
root.title("hello/Goodbye")
root.configure(bg='light blue')
root.geometry('400x400')
    
hello_label = tkinter.Label(root,text='Hello->')
hello_label.configure(bg='light blue')
hello_label.grid(row=0, column=0, padx=5,pady=5,sticky='w')

goodbye_label=tkinter.Label(root,text='Goodbye->')
goodbye_label.configure(bg='light blue')
goodbye_label.grid(row=1, column=0, padx=5,pady=5,sticky='w')

result_button1 = tkinter.Button(root,text='Click me!',command = hello_fun)
result_button1.grid(row=0, column=1)

result_button2 =  tkinter.Button(root,text='Click me!',command = goodbye)
result_button2.grid(row=1, column=1)

root.mainloop()
```

# User input in tkinter

- Another option for user input is to embed a text box directly into the GUI as a widget (as opposed to a popup)
- In tkinter a basic box is called an Entry. For example:
    - new_entry=tkinter.Entry(root,width=4) # 4 is how many characters wide to make the box
    - new_entry.pack()
    
- .get() retunrs what is crrently typed into the box in the string data type
    - It may be importatn to convert this, depending on your application 
- .delete(start,stop) is used to clear the text out of the box
    - This works like string slicing . for example, .delete(3,5) is equivalent to deleting [3:5]
    - Most common use is to clear our the Entry entirely. to do so, use delete(0, tkinter.END)
    
## Communicating beteen widgets
- tkinter allows you to bind widgets to functions in your code. These functions unfortunately may not hae any parameters.
- Since we usually use parameters to communicate with functions, this poses a problem for communicating bewteen widgets
- When communicating bewteen widgets, a solution to this problem is to use **global variables**

```python
import tkinter
import tkinter.messagebox

def close_gui():
    '''Tells the close_gui() function to look at the root variable from outside the function even though it wasnt a parameter'''
    global root
    root.destroy()
root = tkinter.Tk()
close_button=tkinter.Button(root,text='Close', command = close_gui)
close_button.pack()
root.mainloop()
```

```python
import tkinter, tkinter.messagebox

def search():
    """Searches to see if a search string is inside of sport_news.txt."""
    global search_box
    
    with open("sport_news.txt", "r") as file:
        file_contents = file.read()
    
    if search_box.get().lower().strip() in file_contents.lower().strip():
        tkinter.messagebox.showinfo("Success!", "Your search was found.")
    else:
        tkinter.messagebox.showwarning("Failure!", "Your search was not found.")

    search_box.delete(0, tkinter.END)

root = tkinter.Tk()
root.title("Search sport_news.txt")

search_label = tkinter.Label(root, text = "Search:")
search_label.grid(row = 0, column = 0)

search_box = tkinter.Entry(root, width = 10)
search_box.grid(row = 0, column = 1)

search_button = tkinter.Button(root, text = "Submit", command = search)
search_button.grid(row = 0, column = 2)

root.mainloop()
    
```

```python
#import dependencies

import tkinter

import tkinter.messagebox

def SearchTheString():
    """This GUI will search the users input in a pre determined string and tell it if it exists or not"""
    global search_box
    search_string = 'The quick brown fox jumped over the lazy dog.'
    if search_box.get().strip().lower() in search_string.strip().lower():
        tkinter.messagebox.showinfo('Success',"Your search was found")
    else:
        tkinter.messagebox.showwarning('Failure',"Your search was NOT found")
    search_box.delete(0,tkinter.END)


root = tkinter.Tk()

root.title("Searching for your term!")
root.geometry('400x50')

search_label = tkinter.Label(root,text='Search:')

search_label.grid(row=0,column=0)

search_box = tkinter.Entry(root,width=25)

search_box.grid(row=0, column=1)

submit_button=tkinter.Button(root,text='Submit your search', command=SearchTheString)

submit_button.grid(row=0,column=2)

root.mainloop()
```
