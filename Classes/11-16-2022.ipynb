## Sorting

Sort by the values along ether axis 
- df.sort_values('name,ascending=False)

You can sort on multiple dimensions
- df.sort_values(['Lable1','Lable2'],ascending[1,0])

In the above code, the first value will be ascending, and the second will be descending
